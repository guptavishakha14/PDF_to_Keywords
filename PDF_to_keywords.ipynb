```python
from google.colab import files
from pdf2image import convert_from_path
import pytesseract
from indic_transliteration import sanscript
from indic_transliteration.sanscript import transliterate
import nltk
from nltk import FreqDist
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
import re

```

```python
def extract_text_from_pdf():
    uploaded = files.upload()
    pdf_file = next(iter(uploaded))

    # Convert PDF to images
    images = convert_from_path(pdf_file)

    extracted_text = ""

    for image in images:
        # Perform OCR on each image
        text = pytesseract.image_to_string(image, lang='mar')
        extracted_text += text

    return extracted_text
```

```python
def preprocess_text(text):
    # Transliterate text to Devanagari script
    transliterated_text = transliterate(text, sanscript.ITRANS, sanscript.DEVANAGARI)

    # Remove punctuation marks
    text_without_punctuation = re.sub(r'[^\w\s]', '', transliterated_text)

    # Tokenize the text without punctuation
    tokens = nltk.word_tokenize(text_without_punctuation)

    return tokens
```

```python
def extract_features_and_keywords(text):
    # Load Marathi stopwords
    with open('stopwords/marathi.txt', 'r', encoding='utf-8') as file:
        stop_words = [word.strip() for word in file.readlines()]

    # Remove stopwords from the text
    filtered_text = [word for word in text if word.casefold() not in stop_words]

    # Extract features using Part-of-Speech (POS) tagging
    pos_tags = nltk.pos_tag(filtered_text)
    features = [word for word, pos in pos_tags if pos.startswith('N') or pos.startswith('V')]  # Extract nouns and verbs

    # Extract keywords using term frequency
    fdist = FreqDist(filtered_text)
    keywords = [word for word, frequency in fdist.most_common(15)]  # Extract the top 5 most frequent words

    return features, keywords
```

```python
extracted_text = extract_text_from_pdf()
preprocessed_text = preprocess_text(extracted_text)
print("Extracted Text:",extracted_text)
char_count = len(extracted_text)
print("Count of characters:", char_count)
```

```python
extracted_features, extracted_keywords = extract_features_and_keywords(preprocessed_text)
```

```python
print("Extracted Features:", extracted_features)
char_count1 = len(extracted_features)
print("Count of characters:", char_count1)
```
